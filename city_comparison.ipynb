```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('us-cities-table.csv')
```

```python
df
```

```python
df['log_pop'] = np.log(df['pop2024'])
df['log_density'] = np.log(df['densityMi'])
```

```python
durham_row = df[df['city'] == 'Durham']
```

```python
durham_pop = durham_row['log_pop'].iloc[0]
durham_density = durham_row['log_density'].iloc[0]
```

```python
durham_density
```

```python
df['log_pop_diff'] = np.abs(df['log_pop'] - durham_pop)
df['log_density_diff'] = np.abs(df['log_density'] - durham_density)
```

```python
df['log_agg_diff'] = df['log_pop_diff'] + df['log_density_diff']
```

```python
df = df.sort_values(by='log_agg_diff', axis=0)
```

```python
for i, row in df.iterrows():
    print(row['city'])
```
